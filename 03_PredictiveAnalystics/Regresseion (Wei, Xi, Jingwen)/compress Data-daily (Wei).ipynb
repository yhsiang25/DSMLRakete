```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("sf18_weathermerged.csv")
```

We want to reduce our data to 365 rows where each row represents one day with the max temperature, if it rained or not and how many rides were registered on that day.

```python
df.head()
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace= True)
df.head()
```

```python
df["start_time"] = pd.to_datetime(df["start_time"])
```

```python
data = df.groupby(['Date_s']).Date_s.count().reset_index(name = "day_frequency")
data
```

```python
df.info()
```

```python
minsort = df.groupby(['Date_e'])['min_temp'].max().reset_index(name = "maxTemp")
minsort.columns = ['Date_s', 'maxTemp']
#df1.groupby(['State','Product'])['Sales'].max().reset_index()
minsort
```

```python
compress = pd.merge(minsort, data, on='Date_s',  how='left')
compress
```

```python
compress = compress.reindex(['Date_s','day_frequency','maxTemp'], axis=1)
compress
```

```python
def get_date (ts):
    return ts.date()

def get_weekday (ts):
    return ts.weekday()

def get_hour (ts):
    return ts.hour
```

```python
compress["Date_s"] = pd.to_datetime(compress["Date_s"])
compress["weekday"] =compress["Date_s"].apply(lambda x: get_weekday (x))
compress     
```

```python
compress["workingday"] = compress["weekday"].apply(lambda x: 1 if x<=4 else 0)
compress
```

```python
compress['month'] = pd.DatetimeIndex(compress['Date_s']).month
compress
```

```python
rain = df.groupby(['Date_e'])['precip'].max().reset_index(name = "precip")
rain.columns = ['Date_s', 'precip']
#df1.groupby(['State','Product'])['Sales'].max().reset_index()
rain
```

```python
rain.precip = rain.precip.astype(int)
rain
```

```python
rain["Date_s"] = pd.to_datetime(rain["Date_s"])
compress = pd.merge(rain, compress, on='Date_s',  how='left')
compress
```

```python
compress = compress.reindex(['Date_s','day_frequency','maxTemp', 'weekday', 'workingday', 'month', 'precip'], axis=1)
compress
```

```python
compress.to_csv("compressed_by_day.csv")
```
