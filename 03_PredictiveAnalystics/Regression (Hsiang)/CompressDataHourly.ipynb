```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("sf18_weathermerged.csv")
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace= True)
df.head()
```

```python
df["starttime_weather"] = pd.to_datetime(df["starttime_weather"])
```

```python
data = df.groupby(['starttime_weather']).Date_s.count().reset_index(name = "totalRidesPerHour")
data
```

```python
data = data.rename(columns={'starttime_weather':'RideStart'})
data
```

```python
minsort = df.groupby(['starttime_weather'])['min_temp'].max().reset_index(name = "maxTemp")
minsort.columns = ['RideStart', 'maxTemp']
minsort
```

```python
compress = pd.merge(minsort, data, on='RideStart',  how='left')
compress
```

```python
def get_date (ts):
    return ts.date()

def get_weekday (ts):
    return ts.weekday()

def get_hour (ts):
    return ts.hour
```

```python
compress["RideStart"] = pd.to_datetime(compress["RideStart"])
```

```python
compress["weekday"] =compress["RideStart"].apply(lambda x: get_weekday (x))
compress
```

```python
compress["workingday"] = compress["weekday"].apply(lambda x: 1 if x<=4 else 0)
compress
```

```python
compress['month'] = pd.DatetimeIndex(compress['RideStart']).month
compress
```

```python
compress["hour"] =compress["RideStart"].apply(lambda x: get_hour (x))
compress
```

```python
compress["NightTime"] = compress["hour"].apply(lambda x: 1 if x<=4 or x>22 else 0)
compress
```

```python
compress["calendarWeek"] = compress.RideStart.apply(lambda x: x.weekofyear)
compress
```

```python
compress["season"] = compress["month"].apply(lambda x: 1 if x>=3 and x<=5 else 2 if x>=6 and x<=8 else 3 if x>=9 and x<=11 else 4)
compress
```

```python
rain = df.groupby(['starttime_weather'])['precip'].max().reset_index(name = "precip")
rain.columns = ['RideStart', 'precip']
rain
```

```python
rain.precip = rain.precip.astype(int)
rain
```

```python
rain["RideStart"] = pd.to_datetime(rain["RideStart"])
```

```python
compress = pd.merge(compress, rain, on='RideStart',  how='left')
compress
```

```python
compress = compress.fillna(0)
```

```python
compress
```
