```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import json
```

```python
from  folium.plugins import HeatMap
dataset = pd.read_csv("sf18_count_location.csv")
```

```python
dataset.count()
```

```python
dataset.head()
```

```python
data = dataset[['lat', 'lng', 'counts for station_name']].values.tolist()
hmap = folium.Map(location=[37.4700, -122.2500], tiles='stamentoner',
                  control_scale=True, zoom_start=10)
hmap.add_child(HeatMap(data))
file_path = r"test1.html"
hmap.save(file_path)  # save the map as html file
```

## Top 10 and Last 10 station

```python
location_top=dataset.head(10)
location_top
```

```python
location_last=dataset.tail(10)
location_last
```

```python
# Instantiate a feature group for the popular stations in the dataframe
pop_stations = folium.map.FeatureGroup()

# Loop through the 10 stations and add each to the stations feature group
for lat, lng, in zip(location_top.lat, location_top.lng):
    pop_stations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )
    
    
# Instantiate a feature group for the unpopular stations in the dataframe
unpop_stations = folium.map.FeatureGroup()

# Loop through the 10 stations and add each to the stations feature group
for lat, lng, in zip(location_last.lat, location_last.lng):
    pop_stations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='blue',
            fill=True,
            fill_color='purple',
            fill_opacity=0.4
        )
    )
# Add incidents to map
chi_map = folium.Map(location=[37.4700, -122.2500])
chi_map.add_child(pop_stations)
chi_map.add_child(unpop_stations)
```

red circles: most popular stations, purple circles: most unpopular stations
