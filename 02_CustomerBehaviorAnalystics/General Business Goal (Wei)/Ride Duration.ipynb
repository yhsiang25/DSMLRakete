```python
#drop the columns we don not need
num_bike = dataset.drop(["Unnamed: 0","user_type","Date_s","Date_e", "duration [s]","start_station_id", "end_station_id" ,"start_station_name", "Weekday_s", "Weekday_e", "Hour_s", "Hour_e", "end_station_name", "start_station_latitude", "start_station_longitude", "end_station_latitude", "end_station_longitude"], axis =1)
num_bike
```

```python
#see the information of num_bike
num_bike.info()
```

```python
#change the datatype of start- and endtime to datetime  
num_bike["start_time"] = pd.to_datetime(num_bike["start_time"])
num_bike["end_time"] = pd.to_datetime(num_bike["end_time"])
num_bike.info()
```

```python
#insert the month column`
num_bike['month'] = pd.DatetimeIndex(num_bike['start_time']).month
num_bike
```

```python
#drop the columns we don't need
bike_id = num_bike.drop(["start_time","end_time"],axis =1)
bike_id
```

```python
bike_id["bike_id"].value_counts()
```

```python
#drop the repeated bike ids
bike_id2 = bike_id.drop_duplicates()
bike_id2
```

```python
#show the number of bikes in use of everymonth
bikes_counts = bike_id2['month'].value_counts().reset_index(name = 'bike_num' )
bikes_counts.columns = ['month','bike_num']
bikes_counts.sort_values(by='month')
```
