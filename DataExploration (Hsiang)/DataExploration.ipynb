<H3>add a categorical value for the day of the week to be able to access it easier later on.</H3>

```python
# Let's map time of day and day of week to our DF

def get_date (ts):
    return ts.date()

def get_month (ts):
    return ts.month

def get_weekday (ts):
    return ts.weekday()

def get_hour (ts):
    return ts.hour
```

```python
sf18["Month"] = sf18["start_time"].apply(lambda ts: get_month (ts))
sf18["Date_s"] = sf18["start_time"].apply(lambda ts: get_date (ts))
sf18["Weekday_s"]= sf18["start_time"].apply(lambda x: get_weekday (x))
sf18["Hour_s"]= sf18["start_time"].apply(lambda x: get_hour (x))
sf18["Date_e"] = sf18["end_time"].apply(lambda ts: get_date (ts))
sf18["Weekday_e"]= sf18["end_time"].apply(lambda x: get_weekday (x))
sf18["Hour_e"]= sf18["end_time"].apply(lambda x: get_hour (x))
```

<H4>get rid of all data which shows a negative duration and the data with a longer duration than 24h because by BayWheels' policy a bike is marked as stolen and a fine is issued after the ride-time exceeds 24h. </H4>

```python
durationMin = 1 #seconds
durationMax = 86400 #seconds equals 24h -> 
                    #https://help.baywheels.com/hc/en-us/articles/360033793512-Lost-bike-fees 
                    #-> after 24h a bike is registered as stolen

sf18.drop(sf18[sf18["duration [s]"]<durationMin].index, inplace = True)
sf18.drop(sf18[sf18["duration [s]"]>durationMax].index, inplace = True)

sf18
```
