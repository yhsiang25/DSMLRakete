```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime #for working with times objects
from datetime import timedelta #for working with times objects
import math
import random
```

```python
dataset = pd.read_csv("sf18_cleaned.csv")
dataset.head()
```

```python
#drop the columns we don not need
record = dataset.drop(["Unnamed: 0", "bike_id","start_station_id", "end_station_id","duration [s]", "start_station_name", "Weekday_s", "Weekday_e", "Hour_s", "Hour_e", "end_station_name", "start_station_latitude", "start_station_longitude", "end_station_latitude", "end_station_longitude"], axis =1)
record
```

```python
#see the information of record
record.info()
```

```python
#change the datatype of start- and endtime to datetime  
record["start_time"] = pd.to_datetime(record["start_time"])
record["end_time"] = pd.to_datetime(record["end_time"])
record.info()
```

```python
#insert the month column
record['month'] = pd.DatetimeIndex(record['start_time']).month
record
```

## Let's see the amount of users in month

```python
month_sub = record['month'].value_counts().reset_index(name = 'sub_num' )
month_sub.columns = ['month','sub_num']
month_sub.sort_values(by='month')
```

## Let's show the user amount in bar plot

```python
sns.barplot(x="month",y="sub_num",data=month_sub)
```
