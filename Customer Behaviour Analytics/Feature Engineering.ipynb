# 4. Showing the stations in a map

Firstly, we need the station names and how many times they are used

```python
df_times_sum_name_ind_1=df_times_sum_name_ind.set_index(['station_name'])
df_times_sum_name_ind_1
```

Secondly, we need the locations of all the stations

```python
# retain only one row for each station
df_location=sf18[['start_station_name','start_station_latitude','start_station_longitude']]
df_location_unique=df_location.drop_duplicates()
#drop the stations that has a NaN as it's latitude xor longitude
df_location_cleaned=df_location_unique.dropna(axis=0,how='any')

df_location_cleaned_set_index=df_location_cleaned.set_index(['start_station_name'])

df_location_cleaned_set_index=df_location_cleaned_set_index.reset_index()
df_location_cleaned_set_index.columns=['station_name','lat','lng']
df_location_cleaned_set_index
```

Then, merge these two Dataframe

```python
df_location_reset=pd.merge(df_times_sum_name_ind_1, df_location_cleaned_set_index, on = 'station_name').sort_values(by=['counts for station_name'],ascending=False)
df_location_reset
```

```python
df_location_reset.to_csv("sf18_count_location.csv")
```

```python
df_location_top=df_location_reset.head(20)
df_location_top
```

```python
df_location_last=df_location_reset.tail(20)
df_location_last
```

```python
# Instantiate a feature group for the popular stations in the dataframe
pop_stations = folium.map.FeatureGroup()

# Loop through the 10 stations and add each to the stations feature group
for lat, lng, in zip(df_location_top.lat, df_location_top.lng):
    pop_stations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )
    
    
# Instantiate a feature group for the unpopular stations in the dataframe
unpop_stations = folium.map.FeatureGroup()

# Loop through the 10 stations and add each to the stations feature group
for lat, lng, in zip(df_location_last.lat, df_location_last.lng):
    pop_stations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='blue',
            fill=True,
            fill_color='purple',
            fill_opacity=0.4
        )
    )
# Add incidents to map
chi_map = folium.Map(location=[37.4700, -122.2500])
chi_map.add_child(pop_stations)
chi_map.add_child(unpop_stations)
```
